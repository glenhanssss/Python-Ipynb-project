```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dx/dt
def model(x,t):
  x1 = x[0]
  x2 = x[1]
  x3 = x[2]
  x4 = x[3]
  dx1dt = 2*x1+x2 
  dx2dt = 2*x2
  dx3dt = x3-x4
  dx4dt = x3+x4
  dxdt = [dx1dt,dx2dt,dx3dt,dx4dt]
  return dxdt

# initial condition
x0 = [-1,1,2,-3]


# time points
t = np.linspace(0,3)


# solve ODE
x = odeint(model,x0,t)

# plot results
plt.plot(t,x[:,0],'g:',label='x1(t)')
plt.plot(t,x[:,1],'b-',label='x2(t)')
plt.plot(t,x[:,2],'r--',label='x3(t)')
plt.plot(t,x[:,3],'y--',label='x4(t)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend(["x1","x2","x3","x4"])
plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dx/dt
def model(x,t):
  x1 = x[0]
  x2 = x[1]
  x3 = x[2]
  x4 = x[3]
  dx1dt = -2*x1+x2 +x3
  dx2dt = -x1-2*x2 +x4
  dx3dt = -2*x3+x4
  dx4dt = -x3-2*x4
  dxdt = [dx1dt,dx2dt,dx3dt,dx4dt]
  return dxdt

# initial condition
x0 = [1,2,-1,-3]


# time points
t = np.linspace(0,3)


# solve ODE
x = odeint(model,x0,t)

# plot results
plt.plot(t,x[:,0],'g:',label='x1(t)')
plt.plot(t,x[:,1],'b-',label='x2(t)')
plt.plot(t,x[:,2],'r--',label='x3(t)')
plt.plot(t,x[:,3],'y--',label='x4(t)')
plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend(["x1","x2","x3","x4"])
plt.show()
```
